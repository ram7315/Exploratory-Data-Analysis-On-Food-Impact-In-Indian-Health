##### Food plays a important role in India’s culture, economy, and health. Traditional Indian diets, rich in plant-based foods, have historically supported good health. However, rapid urbanization and lifestyle changes have led to a shift towards processed and unhealthy foods, contributing to rising obesity, diabetes, and other non-communicable diseases.

##### At the same time, food insecurity and malnutrition remain significant challenges, particularly in rural areas. Addressing these issues requires promoting healthy eating habits, improving food security, and implementing policies that encourage sustainable and nutritious diets. This analysis focuses on the impact of dietary habits , aiming to identify key factors contributing to its rising prevalence in India.

##### Traditional Indian diets, often plant-based and rich in vegetables, fruits, legumes, and whole grains, have historically been associated with lower rates of chronic diseases. However, the rapid urbanization and economic growth in recent decades have led to a significant shift in dietary patterns. The increased consumption of processed foods, refined sugars, and unhealthy fats has contributed to a rise in obesity, diabetes, and cardiovascular diseases. This dietary transition, combined with sedentary lifestyles, has posed new public health challenges.

<a name="Section51"></a>
<h3 style="color:#78CC01;">8.3 Multi- Varient</h3>

- 1. How do exercise habits and diet type together affect health scores?
- 2. How do spice level, sugar intake, affect BMI?
- 3. Do region, primary cuisine,  affect daily calorie intake?


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Exercise_Level', y='Health_Score', hue='Diet_Type', data=food)
plt.title('Health Score by Exercise Level and Diet Type')
plt.xlabel('Exercise Level')
plt.ylabel('Health Score')
plt.legend(title='Diet Type')
summary = 'This chart shows how health scores vary based on exercise levels (Sedentary, Moderate, Active) and types of diet (Vegetarian, Vegan, Non-Vegetarian).Each colored box represents a diet type.The middle line inside the box shows the average (median) health score.The size of the box shows the range where most of the scores fall.The lines (whiskers) above and below the box show the highest and lowest scores.'
plt.figtext(0.5, -0.1, summary, wrap=True, horizontalalignment='center', fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Spice_Level', y='BMI', hue='Sugar_Intake', data=food)
plt.title('BMI by Spice Level and Sugar Intake')
plt.xlabel('Spice Level')
plt.ylabel('BMI')
plt.legend(title='Sugar Intake')
summary = 'This chart shows the distribution of Body Mass Index (BMI) based on different levels of spice intake (High, Low, Medium) and sugar intake (Moderate, Low, High). Each "violin" shape represents the spread of BMI values. The width of the violin shows how common different BMI values are. Wider areas mean more people have those BMI values.Different colors show different sugar intake levels. The white dot in the middle indicates the median BMI, while the black bar represents the interquartile range (middle 50% of the data)'
plt.figtext(0.5, -0.1, summary, wrap=True, horizontalalignment='center', fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='Region', y='Daily_Calorie_Intake', hue='Primary_Cuisine', data=food)
plt.title('Daily Calorie Intake by Region and Primary Cuisine')
plt.xlabel('Region')
plt.ylabel('Daily Calorie Intake')
plt.legend(title='Primary Cuisine')
summary =' This bar plot shows the average daily calorie intake across regions (East, Central, South, West, North) by cuisine type (Gujarati, North Indian, etc.). Each bars height represents the average intake, with colors indicating different cuisines'
plt.figtext(0.5, -0.1, summary, wrap=True, horizontalalignment='center', fontsize=9)
plt.show()
```
